Mit Hilfe des Python-Moduls pandas lassen sich auch Diagramme darstellen. Die Funktionen die im folgenden betrachtet werden, bekommen als Argumente eine Tabelle und ein Attribut und geben ein Bild zurück.

Die erste Funktion, die betrachtet wird, hat die folgende Struktur:

pie_chart(Tabelle, Attribut)

Zum Beispiel:

pie_chart(animal_table, "Beine")

Wir laden wie im Kapitel 1 zuerst mit Hilfe von pandas den Tierdatensatz:

```python
import pandas as pd # Importiert das Modul pandas unter dem Namen pd
datei_name = "animal_dataset_bc.csv" # Name der Datei
animal_table = pd.read_csv(datei_name, 
                        header=0, # Gibt an, dass die Spaltennamen in der 0. Zeile sind (in der Informatik wir mit 0 zu zählen begonnen)
                        names=["Name", "Spezies", "Geschlecht", "Alter", "Steril", "Beine", "Gewicht", "Wochen"], # Spaltennamen werden gesetzt 
                        index_col=0) # Die 0. Spalte wird für die Labels benutzt
```

```python
from charts import pie_chart # Aus dem Modul charts wird die Funktion pie_chart importiert

pie_chart(animal_table, "Beine") # Erstellt ein Kuchendiagramm
```

**Einzelarbeit:** Was genau zeigt das obige Diagramm? 

*Es zeigt wie viele Tiere wie viele Beine haben. Zum Beispiel haben 3.1% der Tiere 3 Beine.*

Die Funktion pie_chart bekommt als Argument eine Tabelle von Daten (einen Datensatz) und den Namen einer Spalte mit kategorialen Daten und gibt ein **Kuchendiagramm** aus. 

Die Funktion pie_chart geht durch die entsprechende Spalte und zählt, wie oft ein bestimmter Wert vorkommt. Dann wird ein Kuchenstück für den entsprechenden Wert gezeichnet. Das Kuchenstück entspricht dabei der prozentualen Grösse der Anzahl Vorkommnisse des Wertes. 

In diesem Beispiel wurde das animal_table als Tabelle des Datensatzes benutzt und "Beine" (die Anzahl Beine) als Spalte mit kategorialen Daten.

*Balkendiagramme lassen sich oft besser lesen, da Strecken einfacher vergleichbar sind als Winkel. Zudem sind Balken der Höhe 0 besser erkennbar als Stücke der Grösse 0.*

*Gerade wenn viele Kategorien vorhanden sind, sind Kuchendiagramme unübersichtlich und die Kuchenstücke lassen sich von blosem Auge nur schlecht vergleichen. Balkendiagramme haben da Vorteile. Kuchendiagramme sollten nur bei einer kleinen Anzahl Kategorien verwendet werden.* 

*Da Balkendiagramme, zumindest wie sie in diesem Kurs verwendet werden, absolute Häufigkeiten darstellen, eignen sie sich besser zum Vergleich von absoluten Häufigkeiten. Wichtig ist, dass die y-Achse beschriftet ist und mit der 0 beginnt.* 

*Kuchendiagramme haben dafür Vorteile beim Vergleich von relativen Häufigkeiten. Ein Stück ist immer ein Teil vom Ganzen.*
