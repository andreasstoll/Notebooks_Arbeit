Grösste Bibliotheken der Schweiz:

Problem: Whitespaces in den Zahlen

```python
import pandas as pd 
datei_name = "groesste_Bibliotheken_Schweiz_2019.csv"
bibliotheken = pd.read_csv(datei_name,
                            delimiter=";",
                            header=0,
                            names=["Medianangebot", "Ausleihen", "Elektronische Zeitschriften", "EBooks", "Aktive Benutzer"], 
                            index_col=0,
                            )

#Entfernt Leerschläge
for col in bibliotheken.columns:
    bibliotheken[col]=bibliotheken[col].str.replace(" ","")

bibliotheken
```

```python
# Speicher die gereinigte Datei 
bibliotheken.to_csv("Bibliotheken_2019.csv")
```
